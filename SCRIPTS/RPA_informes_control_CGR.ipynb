```python
# Importamos librerias
import pandas as pd
import rpa as r
import time
import os
```

```python
# Leemos el archivo CSV con la información de nombres del comite
df0 = pd.read_excel('/content/drive/MyDrive/DATA HUNTERS - DSRP 2022/1.- BASE DE DATOS/8_Informes_Contraloria/comite_regiones_2019.xls')
df = df0["miembrecomite"]

df = df[1001:2000]
rango = "1001_2000"  # Existen 5000 registros para las 3 regiones

path_export = "/content/drive/MyDrive/DATA HUNTERS - DSRP 2022/1.- BASE DE DATOS/8_Informes_Contraloria/informes_contraloria_"
filetype = ".xlsx"

comite_name = df.values.tolist()
comite_name = list(map(str, comite_name))
```

```python
# Inicializamos RPA
r.init(turbo_mode = True)
```

```python
informe_n_informes = []
informe_name       = []

time.sleep(2)
r.url('https://appbp.contraloria.gob.pe/BuscadorCGR/Informes/Avanzado.html')
time.sleep(6)

 #element_exist = r.present('//*[@id="inpGeneral"]')
 #print(element_exist)

for i in range(len(comite_name)):
    r.type('//*[@id="inpGeneral"]', '[clear]')
    r.type('//*[@id="inpGeneral"]', comite_name[i])
    time.sleep(1)
    r.click('//*[@id="btnBuscar"]')
    time.sleep(1.5)

    element_exist = r.present('//*[@id="tablaResultadosUltimosInformes"]/tbody')

    print(i,element_exist)
    if element_exist == False:
        continue

    n_informes = r.read('//*[@id="lbltotalItems"]')

    informe_n_informes.append(n_informes)
    informe_name.append(comite_name[i])
    time.sleep(0.5)

r.close()

# Creamos el DataFrame
df = pd.DataFrame({ "nombre"       :informe_name, 
                    "n_informes"   :informe_n_informes
                  })

df[['n_informes_control','flag', ]] = df['n_informes'].str.split('|', expand=True)
df['n_informes'] = df['n_informes'].str.slice(6, 8)
inf_contraloria  = df.drop(columns=['n_informes_control', 'flag'])
inf_contraloria.to_excel(path_export+rango+filetype, index = False, header=True)
```
