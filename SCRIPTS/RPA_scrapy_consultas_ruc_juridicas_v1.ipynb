```python
# Leemos el archivo CSV con la información del RUC
df0 = pd.read_excel('/content/drive/MyDrive/DATA HUNTERS - DSRP 2022/1.- BASE DE DATOS/RUCs/rucs_2019_to_sunat_region.xls', 'juridicas')
df = df0["ruc_proveedor_final"]
df = df[1801:2339] #Reemplazar el rango de datos asignado

# 0_2339
rango = "1801_2339"  # Cusco, Piura y MDD

path_export = "/content/drive/MyDrive/DATA HUNTERS - DSRP 2022/1.- BASE DE DATOS/7_SUNAT/sunat_juridicas_scrapy_"
filetype = ".xlsx"

lista_ruc0 = df.values.tolist()
lista_ruc = list(map(str, lista_ruc0))
print(lista_ruc)
```

```python
r.init()
```

```python
sunat_ruc               = []
sunat_tipo_inscripcion  = []
sunat_nombre_comercial  = []
sunat_inscripcion       = []
sunat_estado_contr      = []
sunat_condicion_contr   = []
sunat_direccion         = []
sunat_actividades       = []
sunat_deuda             = []

time.sleep(0.5)
r.url('https://e-consultaruc.sunat.gob.pe/cl-ti-itmrconsruc/jcrS00Alias')
time.sleep(6)

for i in range(len(lista_ruc)):
    r.type('//*[@id="txtRuc"]', '[clear]')
    r.type('//*[@id="txtRuc"]', lista_ruc[i])
    r.click('//*[@id="btnAceptar"]')
    time.sleep(5)

 # SCRAPING de la información que queremos
    element_exist = r.present('/html/body/div/div[2]/div/div[3]/div[2]/div[1]/div/div[2]/h4')
    print(i,element_exist)
    if element_exist == False:
        time.sleep(2)
        r.click('/html/body/div/div[2]/div/div[2]/button')
        continue
    ruc                = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[1]/div/div[2]/h4')
    tipo_inscripcion   = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[2]/div/div[2]/p')
    nombre_comercial   = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[3]/div/div[2]/p')
    inscripcion        = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[4]/div/div[2]')
    estado_contr       = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[5]/div/div[2]/p')
    cond_contr         = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[6]/div/div[2]/p')
    direccion          = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[7]/div/div[2]/p')
    actividades        = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[10]/div/div[2]/table/tbody/tr')
    
    sunat_ruc.append(ruc)
    sunat_tipo_inscripcion.append(tipo_inscripcion)
    sunat_nombre_comercial.append(nombre_comercial)
    sunat_inscripcion.append(inscripcion)
    sunat_estado_contr.append(estado_contr)
    sunat_condicion_contr.append(cond_contr)
    sunat_direccion.append(direccion)
    sunat_actividades.append(actividades)
    time.sleep(1)
    r.click('/html/body/div/div[2]/div/div[2]/button')
    time.sleep(1.1)

r.close() 

# Creamos el DataFrame
sunat = pd.DataFrame({ "ruc"                     :sunat_ruc, 
                       "tipo_inscripcion"        :sunat_tipo_inscripcion,
                       "nombre_comercial"        :sunat_nombre_comercial,
                       "fecha_inscripcion"       :sunat_inscripcion,  
                       "estado_contribuyente"    :sunat_estado_contr, 
                       "condicion_contribuyente" :sunat_condicion_contr, 
                       "direccion"               :sunat_direccion,
                       "actividades"             :sunat_actividades
                     })
sunat.to_excel(path_export+rango+filetype, index = False, header=True)
```
