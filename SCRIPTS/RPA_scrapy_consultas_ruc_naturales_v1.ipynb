```python
# Importamos librerias
import pandas as pd
import rpa as r
import time
import os
```

```python
# Activamos Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Leemos el archivo CSV con la información del RUC
df0 = pd.read_excel('/content/drive/MyDrive/DATA HUNTERS - DSRP 2022/4. SCRIPT/rucs_uniques_request.xls', 'naturales')
df = df0["ruc_proveedor_final"]
df = df[4258:4358] #Reemplazar el rango de datos asignado
rango = "4258_4358"
#rango = "4258_4358"

# Naturales
#[:1192]     Juliana
#[1193:2384] Stalin
#[2385:3576] Ana
#[3577:4771] Maykol

path_export = "/content/drive/MyDrive/DATA HUNTERS - DSRP 2022/4. SCRIPT/sunat_naturales_scrapy_"
filetype = ".xlsx"

lista_ruc0 = df.values.tolist()
lista_ruc = list(map(str, lista_ruc0))
print(lista_ruc)
```

```python
# Inicializamos RPA
r.init(turbo_mode = True)
```

```python
sunat_ruc             = []
sunat_tipo_inscripcion= []
sunat_dni             = []
sunat_inscripcion     = []
sunat_actividades     = []
sunat_estado_contr    = []
sunat_condicion_contr = []
sunat_direccion       = []

time.sleep(2)
r.url('https://e-consultaruc.sunat.gob.pe/cl-ti-itmrconsruc/jcrS00Alias')
time.sleep(6)

for i in range(len(lista_ruc)):
    r.type('//*[@id="txtRuc"]', '[clear]')
    r.type('//*[@id="txtRuc"]', lista_ruc[i])
    time.sleep(1)
    r.click('//*[@id="btnAceptar"]')
    time.sleep(5)

    element_exist = r.present('/html/body/div/div[2]/div/div[3]/div[2]/div[1]/div/div[2]/h4')
    print(i,element_exist)
    if element_exist == False:
        time.sleep(2)
        r.click('/html/body/div/div[2]/div/div[2]/button')
        continue
    ruc              = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[1]/div/div[2]/h4')
    tipo_inscripcion = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[2]/div/div[2]/p')
    dni              = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[3]/div/div[2]/p')
    inscripcion      = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[5]/div/div[2]/p')
    estado_contr     = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[6]/div/div[2]/p')
    condicion_contr  = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[7]/div/div[2]/p')
    direccion        = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[8]/div/div[2]/p')
    actividades      = r.read('/html/body/div/div[2]/div/div[3]/div[2]/div[11]/div/div[2]/table/tbody/tr/td')
  
    sunat_ruc.append(ruc)
    sunat_tipo_inscripcion.append(tipo_inscripcion)
    sunat_dni.append(dni)
    sunat_inscripcion.append(inscripcion)
    sunat_estado_contr.append(estado_contr)
    sunat_condicion_contr.append(condicion_contr)
    sunat_direccion.append(direccion)
    sunat_actividades.append(actividades)
    time.sleep(1)
    r.click('/html/body/div/div[2]/div/div[2]/button')
    time.sleep(2)

r.close()
```

```python
# Creamos el DataFrame
sunat = pd.DataFrame({ "ruc"                     :sunat_ruc, 
                       "tipo_inscripcion"        :sunat_tipo_inscripcion,
                       "dni"                     :sunat_dni,
                       "fecha_inscripcion"       :sunat_inscripcion,  
                       "estado_contribuyente"    :sunat_estado_contr, 
                       "condicion_contribuyente" :sunat_condicion_contr, 
                       "direccion"               :sunat_direccion,
                       "actividades"             :sunat_actividades
                     })
sunat.to_excel(path_export+rango+filetype, index = False, header=True)
```
